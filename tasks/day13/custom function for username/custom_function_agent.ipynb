```python
import os
from dotenv import load_dotenv
import google.generativeai as genai

# Load API key
load_dotenv()
GEMINI_API_KEY = os.getenv("GEMINI_API_KEY")

# Optional if you use Gemini via genai directly
genai.configure(api_key=GEMINI_API_KEY)
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",  
    temperature=0.7,
    google_api_key=GEMINI_API_KEY
)
```

```python
from langchain.agents import Tool
from langchain_experimental.tools.python.tool import PythonREPLTool

def get_user_name(input_text):
    print("\n\nReceived Input:", input_text)
    return "Shafneet Kaur"

tools = [
    Tool.from_function(
        name="GetUserName",
        func=get_user_name,
        description="Returns the user's name."
    )
]
```

```python
from langchain.agents import initialize_agent
from langchain.agents.agent_types import AgentType

agent = initialize_agent(
    tools=tools,
    llm=llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION, # Reasoning + tool use
    verbose=True   
)
```

```python
response = agent.run("what is my name?")
print("\nAnswer:", response)
```
